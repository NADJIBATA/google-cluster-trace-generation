```python
import numpy as np

# Charger les donnes
train = np.load('../data/processed/sequences/train.npy')
val = np.load('../data/processed/sequences/val.npy')

print(f"Train shape: {train.shape}")
print(f"Val shape: {val.shape}")
print(f"Min value: {train.min()}")
print(f"Max value: {train.max()}")
print(f"Mean: {train.mean():.2f}")
print(f"Std: {train.std():.2f}")

# Visualiser quelques squences
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 1, figsize=(12, 8))
for i in range(3):
    axes[i].plot(train[i])
    axes[i].set_title(f'Squence {i}')
    axes[i].grid(True)
plt.tight_layout()
plt.show()
```
